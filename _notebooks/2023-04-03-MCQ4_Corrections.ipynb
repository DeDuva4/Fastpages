# Question 3


The minimum number of connections that need to be broken in order for E to no longer be able to communicate with F is only 3. This can be achieved by breaking the connections between F and A, F and G, and F and E. F-A, F-G, F-E.

# Question 33: Which of the following can be represented by a single binary digit?

I accidentally clicked on the wrong box, but I knew the answer. The remainder when dividing a whole number by two is always going to be zero or 1/2. These two values can be represented in a single binary digit, by having 0 as 0, and 1/2 as 1. The other answer, C, is fairly obvious because the only two possibilities in a boolean variable are True or False. This can be represented in a single binary digit by having False as 0, and True as 1.
